```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/retail_sales_dataset.csv')
print(df.head())
```

```python
print(df.info())
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())

df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Quantity'].fillna(df['Quantity'].mean(), inplace=True)
df['Price per Unit'].fillna(df['Price per Unit'].mean(), inplace=True)
df['Total Amount'].fillna(df['Total Amount'].mean(), inplace=True)

df['Customer ID'].fillna(df['Customer ID'].mode()[0], inplace=True)
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Product Category'].fillna(df['Product Category'].mode()[0], inplace=True)
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Age'] = df['Age'].astype(int)
df['Quantity'] = df['Quantity'].astype(int)
df['Price per Unit'] = df['Price per Unit'].astype(float)
df['Total Amount'] = df['Total Amount'].astype(float)

print(df.info())
```

```python
print(df.head())
```

```python
# Descriptive Statistics: Calculate basic statistics (mean, median, mode, standard deviation)
```

```python
mean_age = df['Age'].mean()
mean_quantity = df['Quantity'].mean()
mean_price_per_unit = df['Price per Unit'].mean()
mean_total_amount = df['Total Amount'].mean()

print(f"Mean Age: {mean_age}")
print(f"Mean Quantity: {mean_quantity}")
print(f"Mean Price per Unit: {mean_price_per_unit}")
print(f"Mean Total Amount: {mean_total_amount}")
```

```python
median_age = df['Age'].median()
median_quantity = df['Quantity'].median()
median_price_per_unit = df['Price per Unit'].median()
median_total_amount = df['Total Amount'].median()

print(f"Median Age: {median_age}")
print(f"Median Quantity: {median_quantity}")
print(f"Median Price per Unit: {median_price_per_unit}")
print(f"Median Total Amount: {median_total_amount}")
```

```python
mode_age = df['Age'].mode()[0]
mode_quantity = df['Quantity'].mode()[0]
mode_price_per_unit = df['Price per Unit'].mode()[0]
mode_total_amount = df['Total Amount'].mode()[0]

print(f"Mode Age: {mode_age}")
print(f"Mode Quantity: {mode_quantity}")
print(f"Mode Price per Unit: {mode_price_per_unit}")
print(f"Mode Total Amount: {mode_total_amount}")
```

```python
std_age = df['Age'].std()
std_quantity = df['Quantity'].std()
std_price_per_unit = df['Price per Unit'].std()
std_total_amount = df['Total Amount'].std()

print(f"Standard Deviation Age: {std_age}")
print(f"Standard Deviation Quantity: {std_quantity}")
print(f"Standard Deviation Price per Unit: {std_price_per_unit}")
print(f"Standard Deviation Total Amount: {std_total_amount}")
```

```python
# Create a summary dataframe
summary_df = pd.DataFrame({
    'Statistic': ['Mean', 'Median', 'Mode', 'Standard Deviation'],
    'Age': [mean_age, median_age, mode_age, std_age],
    'Quantity': [mean_quantity, median_quantity, mode_quantity, std_quantity],
    'Price per Unit': [mean_price_per_unit, median_price_per_unit, mode_price_per_unit, std_price_per_unit],
    'Total Amount': [mean_total_amount, median_total_amount, mode_total_amount, std_total_amount]
})

print(summary_df)
```

```python
# Time Series Analysis: Analyze sales trends over time using time series techniques
```

```python
# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Set the 'Date' column as the index
df.set_index('Date', inplace=True)
```

```python
df['Total Amount'] = df['Quantity'] * df['Price per Unit']

monthly_sales = df['Total Amount'].resample('M').sum()

# Display the monthly sales data
print(monthly_sales.head())
```

```python
import matplotlib.pyplot as plt

# Plot the monthly sales data
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gender distribution
gender_counts = df['Gender'].value_counts()

# Plot gender distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Gender')
plt.title('Gender Distribution')
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Age distribution by gender
plt.figure(figsize=(10, 6))
sns.histplot(df, x='Age', hue='Gender', bins=30, kde=True)
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate total sales
df['Total Amount'] = df['Quantity'] * df['Price per Unit']

# Total sales by gender
sales_by_gender = df.groupby('Gender')['Total Amount'].sum()

# Plot total sales by gender
plt.figure(figsize=(8, 5))
sns.barplot(x=sales_by_gender.index, y=sales_by_gender.values)
plt.title('Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Define age groups
bins = [0, 18, 25, 35, 45, 60, 100]
labels = ['<18', '18-25', '26-35', '36-45', '46-60', '60+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Total sales by age group
sales_by_age_group = df.groupby('Age Group')['Total Amount'].sum()

# Plot total sales by age group
plt.figure(figsize=(10, 6))
sns.barplot(x=sales_by_age_group.index,color='y', y=sales_by_age_group.values)
plt.title('Total Sales by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Most popular product categories
popular_categories = df['Product Category'].value_counts()

# Plot most popular product categories
plt.figure(figsize=(12, 7))
sns.barplot(x=popular_categories.index, y=popular_categories.values,color='g')
plt.title('Most Popular Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Total sales by gender
sales_by_gender = df.groupby('Gender')['Total Amount'].sum().reset_index()

# Plot total sales by gender
plt.figure(figsize=(8, 5))
sns.barplot(x='Gender', y='Total Amount', data=sales_by_gender, palette='viridis')
plt.title('Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Calculate correlation matrix
corr_matrix = df[['Quantity', 'Price per Unit', 'Total Amount']].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Most popular product categories
popular_categories = df['Product Category'].value_counts().reset_index()
popular_categories.columns = ['Product Category', 'Count']

# Plot most popular product categories
plt.figure(figsize=(12, 7))
sns.barplot(x='Product Category', y='Count', data=popular_categories, palette='muted')
plt.title('Most Popular Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Define age groups
bins = [0, 18, 25, 35, 45, 60, 100]
labels = ['<18', '18-25', '26-35', '36-45', '46-60', '60+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Pivot table for heatmap
pivot_table = df.pivot_table(values='Total Amount', index='Age Group', columns='Gender', aggfunc='sum')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt='.1f', cmap='YlGnBu')
plt.title('Sales by Age Group and Gender')
plt.xlabel('Gender')
plt.ylabel('Age Group')
plt.show()
```
